Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sklearn
```

Data Processing

```python
housing = pd.read_csv("AmesHousing.csv")
housing.head()
```

```python
housing.describe()
```

```python
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

housing.hist(bins=50, figsize=(16, 12))
plt.subplots_adjust(hspace=0.5, wspace=0.3)
plt.show()
```

See which data has a lot of null values that can interfere with the test set

```python
print(housing.isnull().sum())
```

Correlation Matrix

```python
corr_matrix = housing.corr(numeric_only=True)
corr_matrix["SalePrice"].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix

attributes = ["Overall Qual", "Gr Liv Area", "Garage Cars", "Garage Area"]

scatter_matrix(housing[attributes], figsize=(12, 8))
plt.show()
```

Because of the high correlation between Garage Cars and Garage Area i decided to only keep the one with the highest correlation, because they measure basically the same thing this means that I keep Garage Cars. 
I also included a few more attributes 

```python
attributes = ["Overall Qual", "Gr Liv Area", "Garage Cars", "Total Bsmt SF", "1st Flr SF", "2nd Flr SF", "Full Bath", "Year Built"]

scatter_matrix(housing[attributes], figsize=(12, 8))
plt.show()
```

```python
print(housing[attributes].isnull().sum())
```

Because only one row is null i could fill it with medians or i could just remove that row. There was also an issue with with there being a lot of correlation between Square Footage, so im gonna try reducing the complexity of the model later by making one attribute containing the total square footage to have better correlation. 

```python
housing["TotalSF"] = housing["Total Bsmt SF"] + housing["1st Flr SF"] + housing["2nd Flr SF"]
features_to_compare = ['SalePrice', 'TotalSF', 'Gr Liv Area', 'Total Bsmt SF', '1st Flr SF', '2nd Flr SF']

print("Correlation with SalePrice:")
print(housing[features_to_compare].corr().loc['SalePrice'].sort_values(ascending=False))
```

Based On the correlation being higher than all the other Square Footage collumns I decided to reduce the complexity and only use a few collumns and using only TotalSF

```python
features = ["TotalSF","Overall Qual", "Garage Cars", "Full Bath", "Year Built"]
print(housing[features].isnull().sum())
```

Since only one collumn has a null data I decided to drop that one instead of replacing it with the mean, because the quantity was so low.

```python
housing.dropna(subset=['Garage Cars'], inplace=True)
housing.dropna(subset=['TotalSF'], inplace=True)
print(housing[features].isnull().sum())

scatter_matrix(housing[features], figsize=(12, 8))
plt.subplots_adjust(hspace=0.3, wspace=0.3)
plt.show()
```

To ensure outputs are almost always the same i decided to use a seed

```python
np.random.seed(42)
```

Creating the test set

```python
from sklearn.model_selection import train_test_split

X = housing[features]
y = housing['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(" Training Data")
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")

print("\n Testing Data ")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")

print(f"\nIs anything null \n{X_test.isnull().sum()}")
```

Since there are no problems with the date we can go forward and start setting up code for a LinearRegression, PolynomialFeatures and an ensemble regression model

Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

train_predictions = lin_reg.predict(X_train)
lin_rmse = np.sqrt(mean_squared_error(y_train, train_predictions))

print(f"Linear Regression Training RMSE: {lin_rmse}")
```

PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

polynomial_regression = Pipeline([
    ("poly_features", PolynomialFeatures(degree=2, include_bias=False)),
    ("lin_reg", LinearRegression()),
])

polynomial_regression.fit(X_train, y_train)

train_predictions_poly = polynomial_regression.predict(X_train)
poly_rmse = np.sqrt(mean_squared_error(y_train, train_predictions_poly))

print(f"Polynomial Regression Training RMSE: {poly_rmse}")
```

Ensemble regression model

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(X_train, y_train)

train_predictions_forest = forest_reg.predict(X_train)
forest_rmse = np.sqrt(mean_squared_error(y_train, train_predictions_forest))

print(f"Random Forest Training RMSE: {forest_rmse}")
```

```python
housing_prepared = housing.copy()

housing_prepared["SalePrice"] = np.log(housing["SalePrice"])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

housing['SalePrice'].hist(ax=axes[0], bins=50)
axes[0].set_title("Original SalePrice Distribution")

housing_prepared['SalePrice'].hist(ax=axes[1], bins=50)
axes[1].set_title("Log-Transformed SalePrice")

plt.show()
```

As we can see there is a right skewed distribution in saleprice so that might have been causing the innacurasies in the model so i will isntead set up a log-transformed model.

```python
X = housing[features]
y = housing_prepared['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(" Training Data")
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")

print("\n Testing Data ")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")

print(f"\nIs anything null \n{X_test.isnull().sum()}")
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

train_predictions = lin_reg.predict(X_train)
lin_rmse = np.sqrt(mean_squared_error(y_train, train_predictions))

print(f"Linear Regression Training RMSE: {lin_rmse}")


polynomial_regression = Pipeline([
    ("poly_features", PolynomialFeatures(degree=2, include_bias=False)),
    ("lin_reg", LinearRegression()),
])

polynomial_regression.fit(X_train, y_train)

train_predictions_poly = polynomial_regression.predict(X_train)
poly_rmse = np.sqrt(mean_squared_error(y_train, train_predictions_poly))

print(f"Polynomial Regression Training RMSE: {poly_rmse}")


forest_reg = RandomForestRegressor()
forest_reg.fit(X_train, y_train)

train_predictions_forest = forest_reg.predict(X_train)
forest_rmse = np.sqrt(mean_squared_error(y_train, train_predictions_forest))

print(f"Random Forest Training RMSE: {forest_rmse}")
```

Final Model Evaluation

```python
final_pred_lin = lin_reg.predict(X_test)
final_pred_poly = polynomial_regression.predict(X_test)
final_pred_forest = forest_reg.predict(X_test)

final_rmse_lin = np.sqrt(mean_squared_error(y_test, final_pred_lin))
final_rmse_poly = np.sqrt(mean_squared_error(y_test, final_pred_poly))
final_rmse_forest = np.sqrt(mean_squared_error(y_test, final_pred_forest))

print("\nFinal Model Evaluation on Test Set")
print(f"Linear Regression Final RMSE: {round(final_rmse_lin,4)}")
print(f"Polynomial Regression Final RMSE: {round(final_rmse_poly,4)}")
print(f"Random Forest Final RMSE: {round(final_rmse_forest,4)}")
```
