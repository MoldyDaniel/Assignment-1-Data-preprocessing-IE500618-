Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Data Processing

```python
housing = pd.read_csv("AmesHousing.csv")
housing.head()
```

```python
housing.info()
```

```python
housing.describe()
```

```python
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

housing.hist(bins=50, figsize=(16, 12))
plt.show()
```

See which data has a lot of null values that can interfere with the test set

```python
print(housing.isnull().sum())
```

Correlation Matrix

```python
corr_matrix = housing.corr(numeric_only=True)
corr_matrix["SalePrice"].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix

attributes = ["Overall Qual", "Gr Liv Area", "Garage Cars", "Garage Area"]

scatter_matrix(housing[attributes], figsize=(12, 8))
plt.show()
```
